Mapping the Values Based on quility Score of the red wine

```python
df = df.replace({
    "quality":{
        8:'Good',
        7:'Good',
        6:'Average',
        5:'Average',
        4:'Bad',
        3:'Bad'
    }
})
df
```

Data Spliting And Scaling 

```python
X = df.drop(columns='quality')
y = df['quality']
scaler = MinMaxScaler(feature_range=(0,1))
X_scaled = scaler.fit_transform(X)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
models={
    "LogisticRegression":LogisticRegression(),
    "SVC":SVC(),
    #"XGBClassifier":XGBClassifier(),
    "CatBoostClassifier":CatBoostClassifier(verbose=False),
    "RandomForestClassifier":RandomForestClassifier(),
    "DecisionTreeClassifier":DecisionTreeClassifier(),
    "KNeighborsClassifier":KNeighborsClassifier(),
    "GaussianNB":GaussianNB(), 
}
```

```python
results = {}
for model_name,model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    results[model_name] = accuracy
    print(f"{model_name} Accuracy:{accuracy}")
    print("\n")
```

```python
model_list = []
accuracy_list = []
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_accuracy = accuracy_score(y_train,y_train_pred)
    test_accuracy = accuracy_score(y_test,y_test_pred)

    accuracy_list.append(test_accuracy)

    model_list.append(list(models.keys())[i])
    print(model)
    print(f"Training Score :{train_accuracy}")
    print(f"Test Score : {test_accuracy}")
    print("="*80)
```

```python
pd.DataFrame(list(zip(model_list,accuracy_list)),columns=['model','score']).sort_values(by=["score"],ascending=False)
```

```python
param_grid ={
    "n_estimators":[100],
    "criterion":["gini", "entropy", "log_loss"],
    "max_depth":[10,15,20],
    "min_samples_split":[2,5,10]
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(),param_grid=param_grid,cv=5,n_jobs=-1,verbose=1)
grid_search.fit(X_train,y_train)
print(f"Best Parameters for Random Forests : {grid_search.best_params_}")
```

```python
best_model = grid_search.best_estimator_
cv_score = cross_val_score(best_model,X_scaled,y,cv = 5)
print(f"Cross Validation Scores for Random Forests :{cv_score}")
print(f"Mean Cross_Validation score :{cv_score.mean()}")
```

```python
# ...existing code...
joblib.dump(best_model, "Best_Model.pkl")
joblib.dump(scaler, "scaler.pkl")

print("Model And Scaler Have been Saved ")
```
